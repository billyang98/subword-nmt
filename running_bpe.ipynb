### getting the vocabulary

```python
from learn_bpe import get_vocabulary
import codecs
```

```python
infile_bpe = codecs.open('../untracked_data_files/all_lyrics_bpe.txt', encoding='utf-8')
```

```python
vocab = get_vocabulary(infile_bpe)
infile_bpe.close()
```

```python
len(vocab.keys())
```

```python
import json
```

```python
bpe_string_token_to_int = {}
int_to_bpe_string_token = {}
vkeys = list(vocab.keys())
for i in range(0, len(vkeys)):
    bpe_string_token_to_int[vkeys[i]] = i + 1
    int_to_bpe_string_token[i + 1] = vkeys[i]
with open('bpe_string_token_to_int.json', 'w') as openfile:
    json.dump(bpe_string_token_to_int, openfile)
with open('int_to_bpe_string_token.json', 'w') as openfile:
    json.dump(int_to_bpe_string_token, openfile)  
with open('vocab_bpe_stats.json', 'w') as openfile:
    json.dump(vocab, openfile)  
```

```python
with open('bpe_string_token_to_int.json') as openfile:
    bpe_string_token_to_int = json.load(openfile)
with open('int_to_bpe_string_token.json') as openfile:
    int_to_bpe_string_token = json.load(openfile)  

for key in bpe_string_token_to_int:
    if str(key) != str(int_to_bpe_string_token[str(bpe_string_token_to_int[str(key)])]):
        print('wrong pairing in bpe_string_token_to_int {}'.format(key))
for key in int_to_bpe_string_token:
    if str(key) != str(bpe_string_token_to_int[str(int_to_bpe_string_token[str(key)])]):
        print('wrong pairing in int_to_bpe_string_token {}'.format(key))
               
```
